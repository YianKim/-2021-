```python
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/Shareddrives/2021민원/데이터/clean_data.csv")
```

1. 분야명과 분야코드

```python
# 분야명=분야코드라서 분야코드 삭제
data = data.drop(["분야코드"], axis=1)
```

2. 등록일시

```python
# 일시 시간 수정
temp=[]
for i in range(len(data['등록일시'])):
  splitdate = str(data['등록일시'][i]).split(' ')
  if len(splitdate)==3:
    temp.append(splitdate[0].split('-') + splitdate[1:3])
  else:
    temp.append(['.','.','.'])

temp=pd.DataFrame(temp)
temp.columns=["연","월","일","오전오후","시간"]

data["월"]=temp["월"]
data["일"]=temp["일"]
data["오전오후"]=temp["오전오후"]
data["시간"]=temp["시간"]

data=data.drop(["등록일시"], axis=1)
```

```python
set(list(data["최상위기관명"]))
# 최상위 기관명에 지역인 것도 있고, 교육청도 있고, 법무부 병무청 등인것도 있고
# 지역에 대한 클러스터링을 통해 지역별 민원 insight를 도출해 보는 것은 어떨까? ex) 경상남도거제
```
